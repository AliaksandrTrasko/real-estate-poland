```python
import pandas as pd

df = pd.read_csv("../data/processed/apartments_all.csv")
```

```python
df.info()
```

```python
df['date_month'] = pd.to_datetime(df['date_month'])
```

```python
print(df['price'].isna().sum())
'''
The original idea was to delete all entries with null in 'price' column.
But because there is 0 null values, there is nothing to delete

'''

total_duplicates = df.duplicated().sum()
print(f"Total dublicates: {total_duplicates}") # 0 dublicates
```

```python
# print(df.isnull().sum()) - shows the number of NaN values in each column
# Filling NaN values in object columns with a placeholder "Unknown"
object_cols = df.select_dtypes(include='object').columns
df[object_cols] = df[object_cols].fillna('Unknown')

df['buildingMaterial'] = df['buildingMaterial'].fillna('Unknown')

#float_cols = df.select_dtypes(include='float64').columns
#df[float_cols] = df[float_cols].fillna
```

```python
# Replacement using dictionary

#1
print(df['type'].unique())

replacements_type = {
    'blockOfFlats': 'Block of Flats',
    'tenement': 'Tenement',
    'apartmentBuilding': 'Apartment Building'
}

df['type'] = df['type'].replace(replacements_type)

#2
print(df['buildingMaterial'].unique())

replacements_material = {
    'concreteSlab': 'Concrete Slab',
    'brick': 'Brick'
}

df['buildingMaterial'] = df['buildingMaterial'].replace(replacements_material)


df = df.rename(columns={
    'squareMeters':        'square_meters',
    'floorCount':          'floor_count',
    'buildYear':           'build_year',
    'centreDistance':      'centre_distance',
    'poiCount':            'poi_count',
    'schoolDistance':      'school_distance',
    'clinicDistance':      'clinic_distance',
    'postOfficeDistance':  'post_office_distance',
    'kindergartenDistance':'kindergarten_distance',
    'restaurantDistance':  'restaurant_distance',
    'collegeDistance':     'college_distance',
    'pharmacyDistance':    'pharmacy_distance',
    'buildingMaterial':    'building_material',
    'hasParkingSpace':     'has_parking_space',
    'hasBalcony':          'has_balcony',
    'hasElevator':         'has_elevator',
    'hasSecurity':         'has_security',
    'hasStorageRoom':      'has_storage_room',
})
```

```python
df.head(-12)    
```

```python
df['rooms'] = df['rooms'].astype('Int64')
df['floor'] = df['floor'].astype('Int64')
df['floor_count'] = df['floor_count'].astype('Int64')
df['build_year'] = df['build_year'].astype('Int64')
```

```python
# Remove spaces at the edges of the text

# Find all object columns
object_cols = df.select_dtypes(include='object').columns

for x in object_cols:
    df[x] = df[x].str.strip() # strip() is used to remove whitespace from the beginning and end of a string
```

```python
# Outlier removal

# Apartments ranging in 10 <= area <= 1,000 sqm
df = df[(df['square_meters'] >= 10) & (df['square_meters'] <= 1000)]

#Price for sale (from 50k to 10 million PLN)
sale_type = (df['offer_type'] == 'sale') & (df['price'] >= 50_000) & (df['price'] <= 10000000)

#Price for sale (from 50 to 50k PLN)
rent_type = (df['offer_type'] == 'rent') & (df['price'] >= 50) & (df['price'] <= 50000)

df = df[sale_type | rent_type]
```

```python
df.to_csv('../data/processed/apartments_all.csv', index=False)
```
