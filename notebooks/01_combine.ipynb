```python
import pandas as pd
from pathlib import Path

data_folder = Path("../data/raw") # current directory is "real_estate poland\notebooks". to get higher to "/data" we use ..
output_path = Path("../data/processed/apartments_all.csv")

# looking for all CSV files in this folder
files = list(data_folder.glob("*.csv"))
    
# Join everything
df = pd.concat([pd.read_csv(f) for f in files], ignore_index=True)

# index=False is needed to prevent Pandas from adding an extra column with row numbers
df.to_csv(output_path, index=False)
```

```python
# 'apartments_pl_*.csv' will find apartments obly like 'apartments_pl_2023_08.csv' - only SALES files
sale_files = list(data_folder.glob("apartments_pl_*.csv"))

# add a ‘offer_type’ column with the value 'sale'
sale_dfs = []
for f in sale_files:
    df = pd.read_csv(f)
    df['offer_type'] = 'sale' 
    df['date_month'] = f.stem.replace('apartments_pl_', '') # This line of code creates a new column called `date_month` in the DataFrame by extracting the date from the filename.
    sale_dfs.append(df)


rent_files = list(data_folder.glob("apartments_rent_pl_*.csv"))

# add a ‘offer_type’ column with the value 'rent'
rent_dfs = []
for f in rent_files:
    df = pd.read_csv(f)
    df['offer_type'] = 'rent' 
    df['date_month'] = f.stem.replace('apartments_rent_pl_', '') # This line of code creates a new column called `date_month` in the DataFrame by extracting the date from the filename.
    rent_dfs.append(df) 
```

```python
# Combine everything (sales and rents with new columns) into one DataFrame
all_dfs = sale_dfs + rent_dfs
final_df = pd.concat(all_dfs, ignore_index=True)

print("Checking the distribution of apartment types:")
print(final_df['offer_type'].value_counts())

# Save in one file
final_df.to_csv(output_path, index=False)
print("Saved")
```
